1. Crie um script Python simples para verifiacar a temperatura de um sistema.
Se for maior que 30, exiba um alerta de superaquecimento.

```python
#Crie um script Python simples para verifiacar a temperatura de um sistema.
# Se for maior que 30, exiba um alerta de superaquecimento.
def verificar_temperatura(temp):
  if temp > 30:
    print('Alerta de superaquecimento!')
  else:
    print('Temperatura normal.')

# Solicita a temperatura ao usuário
temperatura_digitada = int(input('Digite a temperatura atual do sistema: '))
verificar_temperatura(temperatura_digitada)
```

2. Ciar um programa que organize arquivos bagunçados em pastas automaticamente, baseando-se apenas na extensão do arquivo (ex. .pdf, .jpg)

```python
import os
import shutil

def organize_files_by_extension(directory_path='.'):
  """Organizes files in a given directory into subfolders based on their extensions."""
  if not os.path.isdir(directory_path):
    print(f"Erro: O diretório '{directory_path}' não existe.")
    return

  print(f"Organizando arquivos em: {os.path.abspath(directory_path)}")

  for filename in os.listdir(directory_path):
    if os.path.isfile(os.path.join(directory_path, filename)):
      file_extension = os.path.splitext(filename)[1].lower()

      if not file_extension: # handle files without extension
        folder_name = 'sem_extensao'
      else:
        folder_name = file_extension[1:] # remove the dot

      destination_folder = os.path.join(directory_path, folder_name)

      if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
        print(f"Criado diretório: {destination_folder}")

      try:
        shutil.move(os.path.join(directory_path, filename), destination_folder)
        print(f"Movido: '{filename}' para '{destination_folder}'")
      except shutil.Error as e:
        print(f"Erro ao mover '{filename}': {e}")
  print("Organização concluída!")

# Example usage:
# First, let's create some dummy files for demonstration
# You can change the 'test_directory' to the path you want to organize.
test_directory = 'arquivos_para_organizar'
if not os.path.exists(test_directory):
    os.makedirs(test_directory)

# Create some dummy files in the test directory
open(os.path.join(test_directory, 'documento1.pdf'), 'a').close()
open(os.path.join(test_directory, 'imagem1.jpg'), 'a').close()
open(os.path.join(test_directory, 'documento2.txt'), 'a').close()
open(os.path.join(test_directory, 'imagem2.png'), 'a').close()
open(os.path.join(test_directory, 'arquivo_sem_extensao'), 'a').close()
open(os.path.join(test_directory, 'relatorio.pdf'), 'a').close()
open(os.path.join(test_directory, 'musica.mp3'), 'a').close()

print(f"Arquivos dummy criados em '{test_directory}'.")

# Now, call the function to organize them
organize_files_by_extension(test_directory)

# To organize files in the current working directory, you would call:
# organize_files_by_extension()
```
