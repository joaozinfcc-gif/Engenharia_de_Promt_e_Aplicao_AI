### 1. Requisição GET e Manipulação de JSON

Primeiro, vamos fazer uma requisição GET para uma API pública que retorna dados em formato JSON (usaremos o JSONPlaceholder para exemplos de posts). O objetivo é obter os dados e mostrá-los.

```python
import requests
import json

# URL da API para posts de exemplo
api_url = 'https://jsonplaceholder.typicode.com/posts/1'

# Realiza a requisição GET
response = requests.get(api_url)

# Verifica se a requisição foi bem-sucedida (código de status 200)
if response.status_code == 200:
    # Converte o payload JSON da resposta em um dicionário Python
    data = response.json()

    print("Requisição GET bem-sucedida!")
    print("Tipo de dado recebido:", type(data))
    print("Payload JSON (primeiro post):")
    display(data)

    # Exemplo de manipulação do JSON: acessar um campo específico
    print("\nTítulo do post:", data['title'])
    print("Corpo do post:", data['body'][:50], "...") # Exibe os primeiros 50 caracteres do corpo
else:
    print(f"Erro na requisição GET: {response.status_code}")
```

### 2. Requisição POST com Payload JSON

Agora, vamos realizar uma requisição POST para enviar dados em formato JSON para a API. Geralmente, as requisições POST são usadas para criar novos recursos. A API de exemplo retornará o objeto criado com um ID simulado.

```python
import requests
import json

# URL da API para criar um novo post
api_url = 'https://jsonplaceholder.typicode.com/posts'

# Dados a serem enviados no payload JSON
new_post_payload = {
    'title': 'Meu Novo Post via Python',
    'body': 'Este é o corpo do meu post que estou enviando usando a biblioteca requests em Python.',
    'userId': 1
}

# Define o cabeçalho Content-Type para indicar que estamos enviando JSON
headers = {'Content-Type': 'application/json'}

# Realiza a requisição POST
# O parâmetro 'json' do requests já cuida da serialização do dicionário para string JSON
response = requests.post(api_url, json=new_post_payload, headers=headers)

# Verifica se a requisição foi bem-sucedida (código de status 201 para 'Created')
if response.status_code == 201:
    # Converte o payload JSON da resposta em um dicionário Python
    created_data = response.json()

    print("Requisição POST bem-sucedida! Recurso criado:")
    display(created_data)

    print("\nNovo ID do post:", created_data['id'])
else:
    print(f"Erro na requisição POST: {response.status_code}")
    print("Resposta da API:", response.text)
```

Este é um boilerplate básico para realizar requisições HTTP e trabalhar com JSON. Você pode expandir este código para diferentes tipos de requisições (PUT, DELETE) e para lidar com autenticação, tratamento de erros mais robusto, etc.
